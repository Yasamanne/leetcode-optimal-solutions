```python
######## QUESTION ###########
'''
Leetcode #49

Given an array of strings strs, group the anagrams together. You can return the answer in any order.

An Anagram is a word or phrase formed by rearranging the letters of a different word or phrase, typically using all the original letters exactly once.

 

Example 1:

Input: strs = ["eat","tea","tan","ate","nat","bat"]
Output: [["bat"],["nat","tan"],["ate","eat","tea"]]
Example 2:

Input: strs = [""]
Output: [[""]]
Example 3:

Input: strs = ["a"]
Output: [["a"]]
 

Constraints:

1 <= strs.length <= 104
0 <= strs[i].length <= 100
strs[i] consists of lowercase English letters.

'''

###### Questions to ask interviewr #####
'''
1. Are there any capital letters?
2. Should each return array be sorted alphabetically?

'''


######## Solution ########

class Solution:
    def groupAnagrams(self, strs: List[str]) -> List[List[str]]:

        if not strs:
            return
        m = defaultdict(list)
        for str in strs:
            key = ''.join(sorted(str))
            m[key].append(key)
        return m.values()
       
    
## Time Complexity: O(n.k.log(k)) => k is the length of longest word in strs
## Space Complexity: O(n.k) => sorting requires o(k) space for a string legth of k
```
